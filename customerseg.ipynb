```python
AgeGroupsdf = df1[['Age','Spending_Score']].copy()
```

```python
def AgeRange(value):
	if value < 23:
		return "18-22"
	if 23 <= value < 28:
		return "23-27"
	 elif 28 <= value < 36:
		return "28-35"
	 elif 36 <= value < 46:
		return "36-45"
   elif 46 <= value < 56:
		return "46-55"
   elif 56 <= value < 66:
		return "56-65"
	 elif value >= 90:
		return "65-90"

AgeGroupsdf['AgeRnge'] = AgeGroupsdf['Age'].map(AgeRange)
display(AgeGroupsdf.head())
```
